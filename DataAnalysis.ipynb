```python
# Calcula distribucion de distancias y genera adjacencias

r0 = 4.0

fig = plt.figure(figsize=(15,15), dpi = 300)

t = 100 #paso de tiempo

ruta_dist = path + "dists/" + repr(t) + ".txt"
dists = open(ruta_dist, 'rb')

num_bin = 150.

bins = arange(1,num_bin+1)

DD = []

hist = zeros(num_bin)

#print hist

j = 0

for line in dists.readlines():
    
    vals = line.split()
    
    for k in range(len(vals)-j):
        
        DD.append(float(vals[k]))
        
    j += 1
        
DD.sort()

#print DD[0]
#print len(DD)

epsilon = DD[-1]/num_bin

for dist in DD:
    
    b = dist/epsilon
    
    if b>0: hist[int(b-1)] += 1.
    
#print hist    

#print DD[-1]
#print len(DD)
#print DD[0]
#print DD[len(DD)-1]

#print bins
#print epsilon*bins

plt.plot(epsilon*bins,hist)
plt.title('Distribucion de distancias');
plt.grid(True)


#print epsilon
```

```python
#print R

edges = []
tiempo = []
coef = []
conect = []

G = nx.Graph()

for i in range (int(R)):
    
    #fig = plt.figure(figsize=(10,10), dpi = 300)

    AL = open(path+"/adj_mat/"+ str(i*250) +".dat", 'rb')

    #G = nx.read_adjlist(AL,'#' ,'\t', nodetype = int)
    G = nx.read_adjlist(AL,'#','\t', nodetype = int, create_using = nx.Graph())
    
    pos = nx.random_layout(G, 100)
    
    #nx.draw_random(G)
    #nx.draw_circular(G)
    
    coef.append(nx.average_clustering(G))
    
    conect.append(nx.number_connected_components(G))
        
    edges.append(G.number_of_edges())
    ###edges.append(G.size())
    tiempo.append(i*250)

    #print G.number_of_edges()
    
#print edges
#print tiempo

#fig = plt.figure(figsize=(15,10), dpi = 300)

fig, axes = plt.subplots(nrows = 3, ncols = 1, figsize = (15,35), dpi = 300)
#fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (15,35), dpi = 500)

axes[0].plot(tiempo,edges)
axes[0].grid(True)
axes[0].set_title("Numero de enlaces")
axes[0].set_xlabel("Tiempo")

axes[1].plot(tiempo,coef)
axes[1].grid(True)
axes[1].set_title("Coeficiente clustering")
axes[1].set_xlabel("Tiempo")

axes[2].plot(tiempo,conect)
axes[2].grid(True)
axes[2].set_title("Numero componentes conectados")
axes[2].set_xlabel("Tiempo")
```
