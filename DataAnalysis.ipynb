```python
fig, axes = plt.subplots(nrows = 4, ncols = 1, figsize = (10,35), dpi = 300)
#fig = plt.figure(figsize=(10,10), dpi = 300)

num_bin = 150.0
paso = 250

bins = arange(1,num_bin+1)

R_PROM = []
R_MAX = []
STD_DEV = []
tiempo = []

#for i in range(R/2):
for i in range(1,int(R)):    

    ruta_dist = path + "dists/" + repr(i*paso) + ".txt" #Ruta a archivos de matriz de distancias
    dists = open(ruta_dist, 'rb') #Abre el archivo para leer
    
    tiempo.append(i*paso)
    
    DD = []
    
    hist = zeros(num_bin)

    #print hist

    j = 0

    for line in dists.readlines():

        vals = line.split()

        for k in range(len(vals)-j):

            DD.append(float(vals[k]))

        j += 1

    DD.sort()

    #print DD[0]
    #print len(DD)

    epsilon = DD[-1]/num_bin

    for dist in DD:

        b = dist/epsilon

        if b>0: hist[int(b-1)] += 1.
        
    r_prom = 0
    
    for i in range(int(num_bin)):
        
        r_prom += epsilon*(i+1)*hist[i]
        
    R_PROM.append(r_prom/float(len(DD)))
    
    #print hist.index(max(hist))
    #print np.argmax(hist)
    
    R_MAX.append(np.argmax(hist)*epsilon)
    
    STD_DEV.append(np.std(DD))

    #print hist    

    #print DD[-1]
    #print len(DD)
    #print DD[0]
    #print DD[len(DD)-1]

    axes[0].plot(epsilon*bins,hist)
    #plt.plot(epsilon*bins,hist)
    
axes[0].grid(True)
axes[0].set_title('Distribucion de distancias')


#plt.grid(True)
#plt.title("Distribución de distancias")
#plt.xlabel("Tiempo")
#fig.savefig("dists1.png", dpi=500) 

axes[1].plot(tiempo,R_PROM)
axes[1].grid(True)
axes[1].set_title("Distancia promedio")

axes[2].plot(tiempo,R_MAX)
axes[2].grid(True)
axes[2].set_title("Distancia mas probable")

axes[3].plot(tiempo,STD_DEV)
axes[3].grid(True)
axes[3].set_title("Desviacion estandar")

#print epsilon
```
